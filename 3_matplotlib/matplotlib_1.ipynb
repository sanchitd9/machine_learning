# Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
x
```

```python
y
```

### Creating a simple plot

```python
plt.plot(x, y)
plt.xlabel("X Label")
plt.ylabel("Y Label")
plt.title("Title")
```

### Creating sub-plots

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, "r")

plt.subplot(1, 2, 2)
plt.plot(y, x, "b")
```

### Setting axes' labels and the title

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x, y)
axes.set_xlabel("X Label")
axes.set_ylabel("Y Label")
axes.set_title("Set Title")
```

### Creating a plot within a plot

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes1.plot(x, y)
axes1.set_title("LARGER_PLOT")

axes2.plot(y, x)
axes2.set_title("SMALLER PLOT")
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x, y)
axes[0].set_title("First Plot")
axes[1].plot(y, x)
axes[1].set_title("Second Plot")
```

### Setting the size of the plot

```python
fig = plt.figure(figsize=(3, 2))
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y)
```

### Sub-plots with a custom size

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(3, 4))
axes[0].plot(x, y)
axes[1].plot(y, x)
plt.tight_layout()
```

### Saving a plot

```python
fig.savefig("my_picture.png")
```

### Adding a legend

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, x ** 2, label="X Squared")
ax.plot(x, x ** 3, label="X Cubed")

ax.legend(loc=(0.1, 0.1))
```

### Setting various plot parameters like marker size, marker style, line style, etc.

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y, color="purple", linewidth=3, alpha=0.5, linestyle="-", marker="o", markersize=10, markerfacecolor="yellow", markeredgewidth=3, markeredgecolor="green")
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y, color="purple", linewidth=2, ls="--")

ax.set_xlim([0, 1])
ax.set_ylim([0, 2])
```
