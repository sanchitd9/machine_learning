# Seaborn

```python
import seaborn as sns
```

```python
%matplotlib inline
```

### Load a dataset

```python
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

### Histogram

```python
sns.displot(tips["total_bill"], bins=40)
```

### Jointplot - Scatter + Histogram

```python
sns.jointplot(x="total_bill", y="tip", data=tips)
```

### Pairplot - Scatter plot matrix of all descriptive features

```python
sns.pairplot(tips, hue="sex")
```

### Rug plot

```python
sns.rugplot(tips["total_bill"])
```
